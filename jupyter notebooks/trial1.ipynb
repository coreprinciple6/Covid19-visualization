```python
data= pd.read_csv('datasets/C19_dataset.csv')
data.head()
```

```python
def marker(data):
    # generate a new map
    folium_map = folium.Map(location=[23.777176, 90.399452],
                            zoom_start=8,
                            tiles="CartoDB dark_matter",
                            width='100%')

    # for each row in the data, add a cicle marker
    for index, row in data.iterrows():
        
        # generate the popup message that is shown on click.
        popup_text = "{}<br>cases: {}<br>Deaths: {}"
        popup_text = popup_text.format(row["District/City"],
                          row["Total"],
                          row["Total Death"])
                    
        
        # radius of circles
        radius = 8
        
        # choose the color of the marker
        color="#0A8A9F"
        
        # add marker to the map
        folium.CircleMarker(location=(row["Latitude"],
                                      row["Longitude"]),
                            radius=radius,
                            color=color,
                            popup=popup_text,
                            fill=True).add_to(folium_map)
    return folium_map
```

```python
folium_map = marker(data)
folium_map
```

```python
folium_map.save("final.html")
```
