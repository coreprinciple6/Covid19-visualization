```python
data= pd.read_csv('../dataset/C19_dataset.csv')  # updated till may 6th
data.head()
```

```python
from folium.features import DivIcon
def marker(data):
    # generate a new map
    folium_map = folium.Map(location=[23.777176, 90.399452],
                            zoom_start=8,
                            tiles="CartoDB dark_matter",
                            width='100%')

    # for each row in the data, add a cicle marker
    for index, row in data.iterrows():
              
        # radius of circles
        radius = pow(row["Total"],0.4)
        
        # choose the color of the marker
        color="#BEF243"
        
        # toolip
        tool = "{}<br>cases: {}<br>Deaths: {}"
        tool=tool.format(row["District/City"],
                          row["Total"],
                          row["Total Death"]),
        
        # add marker to the map
        folium.CircleMarker(location=(row["Latitude"],
                                      row["Longitude"]),
                            radius=radius,
                            color= color,
                            icon=folium.DivIcon,
                            tooltip=tool,
                            fill=True).add_to(folium_map)
     
    return folium_map
```

```python
folium_map = marker(data)
folium_map
```

```python
folium_map.save("updated3.html")
```
